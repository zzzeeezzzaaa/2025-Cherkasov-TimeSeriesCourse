Чтобы разобраться, почему так происходит и как работает данный алгоритм, ниже мы реализуем все этапы алгоритма DRAG.

Как мы помним из лекций алгоритм DRAG содержит два этапа:

1. Отбор - За одно сканирование ряда сформировать множество кандидатов в диссонансы.
2. Очистка - За одно сканирование ряда отбросить кандидатов, которые являются ложными диссонансами.

#### **3.3 Поиск диссонансов с помощью алгоритма Merlin**

Условно мы можем разделить поиск диссонансов на три шага:

1. Поиск диссонансов минимальной длинны $minL$. На данном шаге $r = 2\sqrt{minL}$.
2. Поиск диссонансов следующих четырех длин. На данном шаге $r = 0.99 \cdot nndist_{m-1}$. Где $nndist_{m-1}$ - расстояние до ближайшего соседа предыдущего найденного диссонанса.
3. Поиск диссонансов всех
оставшихся дли. $r = \mu - 2 \sigma$. Средние значение и стандартное отклонение вычисляются из расстояний 5 предыдущих диссонансов. н

##### 3.3.1 Поиск диссонансов минимальной длинны

Вспомнил последовательность действий первого шага алгоритма:
![merlin-part-first](pics/first_part.png)

В данной работе мы внесем небольшое изменение, мы будем считать, что подпоследовательность может быть диссонансом только в том случае, если больше 75% точек, не входят в состав других диссонансов.

```python
T = walk_run
m = 50
# сформируем массив метод для потенциальных кандидатов в диссонансы.
# после каждого найденного диссонанса,
# мы будем исключать окружающие его подпоследовательности из числа потенциальных кандидатов,
# путем замены значений их меток на false
excl_zone = int(np.ceil(m / 4))
include = np.ones(len(T)-m+1, dtype=bool)
# Количество диссонансов, которые мы будем искать
topK = 10
```

```python
dis_idx = -np.ones((topK))
dis_nnDist = -np.ones((topK))
dis_nn_idx = np.full((topK),-np.inf)
#первое прближение r
r = 2*np.sqrt(m)
minL = m
maxL = int(m+np.ceil(m*0.1))
#количество найденных диссонасов
cound_find_dis = 0

while dis_nnDist[cound_find_dis-1]<0 and cound_find_dis<topK:
    result = DRAG(data=T,m=minL,r=r, include =include)
    for diss, nnDist, nn in zip(*result):
        dis_idx[cound_find_dis] = diss
        dis_nnDist[cound_find_dis] = nnDist
        dis_nn_idx[cound_find_dis] = nn
        #исключаем окружающие найденный диссонас
        #подпоследовательности и числа потенциальных диссонасов
        core.apply_exclusion_zone(include, diss, excl_zone, False)
        cound_find_dis+=1
        if cound_find_dis>=topK:
            break
    r*=0.5
```

Реализуйте, оставшиеся шаги алгоритма, чтобы найти оставшиеся диссонансы.

```python
import numpy as np

# Исходные данные и базовые параметры
time_series = walk_run  # анализируемый временной ряд
min_subsequence_length = 50  # минимальная длина подпоследовательности для анализа
exclusion_zone_radius = int(np.ceil(min_subsequence_length / 4))  # радиус зоны исключения вокруг найденных диссонансов
top_k_dissonances = 10  # количество диссонансов для поиска

# Массивы для хранения информации о найденных диссонансах
dissonance_positions = -np.ones((top_k_dissonances))  # позиции диссонансов в ряду
dissonance_distances = -np.ones((top_k_dissonances))  # расстояния до ближайших соседей
dissonance_neighbors = np.full((top_k_dissonances), -np.inf)  # индексы ближайших соседей

search_radius = 2 * np.sqrt(min_subsequence_length)  # начальный радиус поиска
current_length = min_subsequence_length  # текущая длина анализируемых подпоследовательностей
max_subsequence_length = int(min_subsequence_length + np.ceil(min_subsequence_length * 0.1))  # максимальная длина
found_dissonances_count = 0  # счетчик найденных диссонансов

# Маска для исключения уже обработанных участков временного ряда
included_positions = np.ones(len(time_series) - min_subsequence_length + 1, dtype=bool)

# ЭТАП 1: ПЕРВИЧНЫЙ ПОИСК С ФИКСИРОВАННОЙ ДЛИНОЙ

print("=== ЭТАП 1: Поиск с фиксированной длиной подпоследовательностей ===")

# Поиск диссонансов с минимальной длиной подпоследовательностей
while dissonance_distances[found_dissonances_count - 1] < 0 and found_dissonances_count < top_k_dissonances:
    # Выполняем поиск диссонансов с текущими параметрами
    search_result = DRAG(data=time_series, m=current_length, r=search_radius, include=included_positions)
    
    # Обрабатываем найденные диссонансы
    for dissonance_pos, neighbor_distance, neighbor_idx in zip(*search_result):
        # Сохраняем информацию о найденном диссонансе
        dissonance_positions[found_dissonances_count] = dissonance_pos
        dissonance_distances[found_dissonances_count] = neighbor_distance
        dissonance_neighbors[found_dissonances_count] = neighbor_idx
        
        # Исключаем область вокруг найденного диссонанса из дальнейшего поиска
        core.apply_exclusion_zone(included_positions, dissonance_pos, exclusion_zone_radius, False)
        
        found_dissonances_count += 1
        if found_dissonances_count >= top_k_dissonances:
            break
    
    # Уменьшаем радиус поиска для следующей итерации
    search_radius *= 0.5

# ЭТАП 2: ПОИСК С ПОСТЕПЕННЫМ УВЕЛИЧЕНИЕМ ДЛИНЫ (ШАГ 1)

print("=== ЭТАП 2: Поиск с постепенным увеличением длины ===")

# Поиск с небольшим увеличением длины подпоследовательностей
for current_length in range(min_subsequence_length + 1, min_subsequence_length + 5):
    if found_dissonances_count >= top_k_dissonances:
        break
        
    # Устанавливаем радиус поиска на основе предыдущих результатов
    search_radius = 0.99 * dissonance_distances[found_dissonances_count - 1]
    
    while dissonance_distances[found_dissonances_count - 1] < 0 and found_dissonances_count < top_k_dissonances:
        # Поиск с увеличенной длиной подпоследовательностей
        search_result = DRAG(data=time_series, m=current_length, r=search_radius, include=included_positions)
        
        for dissonance_pos, neighbor_distance, neighbor_idx in zip(*search_result):
            dissonance_positions[found_dissonances_count] = dissonance_pos
            dissonance_distances[found_dissonances_count] = neighbor_distance
            dissonance_neighbors[found_dissonances_count] = neighbor_idx
            
            core.apply_exclusion_zone(included_positions, dissonance_pos, exclusion_zone_radius, False)
            
            found_dissonances_count += 1
            if found_dissonances_count >= top_k_dissonances:
                break
        
        # Постепенно уменьшаем радиус поиска
        search_radius *= 0.99

# ЭТАП 3: АДАПТИВНЫЙ ПОИСК С ДИНАМИЧЕСКИМ РАДИУСОМ

print("=== ЭТАП 3: Адаптивный поиск с динамическим радиусом ===")

# Поиск с дальнейшим увеличением длины и адаптивным радиусом
for current_length in range(min_subsequence_length + 5, max_subsequence_length + 1):
    if found_dissonances_count >= top_k_dissonances:
        break
    
    # Вычисляем адаптивный радиус поиска на основе статистики предыдущих результатов
    recent_distances = dissonance_distances[max(0, found_dissonances_count - 5):found_dissonances_count]
    mean_distance = np.mean(recent_distances)  # среднее расстояние последних диссонансов
    std_distance = np.std(recent_distances)    # стандартное отклонение расстояний
    
    # Устанавливаем радиус как среднее минус два стандартных отклонения
    search_radius = mean_distance - 2 * std_distance

    while dissonance_distances[found_dissonances_count - 1] < 0 and found_dissonances_count < top_k_dissonances:
        search_result = DRAG(data=time_series, m=current_length, r=search_radius, include=included_positions)
        
        for dissonance_pos, neighbor_distance, neighbor_idx in zip(*search_result):
            dissonance_positions[found_dissonances_count] = dissonance_pos
            dissonance_distances[found_dissonances_count] = neighbor_distance
            dissonance_neighbors[found_dissonances_count] = neighbor_idx
            
            core.apply_exclusion_zone(included_positions, dissonance_pos, exclusion_zone_radius, False)
            
            found_dissonances_count += 1
            if found_dissonances_count >= top_k_dissonances:
                break
        
        # Уменьшаем радиус поиска на величину стандартного отклонения
        search_radius -= std_distance

# ВЫВОД РЕЗУЛЬТАТОВ
print(f"Найденные диссонансы: {dissonance_positions}")
print("Расстояния до ближайших соседей:", dissonance_distances)
```
